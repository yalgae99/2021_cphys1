# 1. Leibniz Pi formula

\begin{align}
\frac{\pi}{4} = \frac{1}{1}-\frac{1}{3}+\frac{1}{5} -\frac{1}{7} + \frac{1}{9} - \cdots 
\end{align}
이 식은 라이프니츠의 파이 공식이다.  
1. 다음의 식을 구현하는 함수를 만들어라.  
\begin{align}
f(n) = \frac{1}{1}-\frac{1}{3}+\frac{1}{5} -\frac{1}{7} + \frac{1}{9} - \cdots + \frac{1}{n}.
\end{align}
2. 이 식을 이용하여 아래처럼 그림을 그려라.  

# 2. Logistic Map

로지스틱 맵은 다음처럼 주어진다.
$$ f_{n+1} = r f_n (1- f_n). $$  

1. $f_n$을 구하는 함수를 작성하여라.
2. 작성한 함수를 이용하여 다음의 조건에서 $n=1$에서 $100$까지 로지스틱 맵의 값을 구하고 그래프로 그려라. 즉, x축은 $n$, y축은 $f_n$인 그래프를 그려라.
  * $r=0.5$, $f_0$=0.5
  * $r=0.95$, $f_0$=0.5
  * $r=2$, $f_0$=0.5
  * $r=3.2$, $f_0$=0.5
  * $r=3.8$, $f_0$=0.5
  * $r=3.8$, $f_0$=0.501

# 3. sinc function

1. 싱크 함수를 그려보아라. 싱크함수는 다음처럼 주어진다.
$$ sinc(x)  = \frac{\sin x}{x}. $$
2. 싱크 함수를 테일러 전개하면 다음과 같다.
$$ sinc(x)  = \frac{sin x}{x} = \sum_{n=0}^{\infty} \frac{(-1)^n x^{2n} }{(2n+1)!}. $$ 
위의 식의 우변을 $n$을 바꾸어 가면서  그래프로 그리고 싱크함수와 비교하여라. 

```python
#1 Leibniz Pi formula

def leibniz(n):
    sum=0
    i=0
    while (2*i+1)<=n:
        sum+=(-1)**i/(2*i+1)
        i+=1
    return sum
n=np.arange(1,100,2)
plt.plot(n,[leibniz(i) for i in n],label='Leibniz formula for $\pi$')
plt.plot(n,[np.pi/4 for i in range(50)],label='$\pi$/4')
plt.legend()
plt.xlabel('n')
plt.show()
```

```python
#2 Logistic map
def logistic(n,r,f0):
    if n==0:
        return f0
    else:
        f=f0
        for i in range(n):
            f=r*f*(1-f)
        return f
n1=np.arange(1,101,1)
plt.plot(n1,[logistic(i,0.5,0.5) for i in n1])
plt.plot(n1,[logistic(i,0.95,0.5) for i in n1])
plt.plot(n1,[logistic(i,2,0.5) for i in n1])
plt.plot(n1,[logistic(i,3.2,0.5) for i in n1])
plt.plot(n1,[logistic(i,3.8,0.5) for i in n1])
plt.plot(n1,[logistic(i,3.8,0.501) for i in n1])
plt.show()
```

```python
#3 unnormalized sinc function
    
def sinc(x):
    return np.sinc(x/np.pi)
def sinc_tn(x,N):
    j=0
    sum=0
    while j<=N:
        if j%2==0:
            sum+=(x**(2*j))/np.math.factorial(2*j+1)
            j=j+1
        else:
            sum-=(x**(2*j))/np.math.factorial(2*j+1)
            j=j+1
    return sum
```

```python
n2=np.arange(1,10,0.001)
plt.plot(n2,[sinc(i) for i in n2],label='sinc(x)')
plt.plot(n2,[sinc_tn(x,9) for x in n2],label='N=9')
plt.plot(n2,[sinc_tn(x,10) for x in n2],label='N=10')
plt.plot(n2,[sinc_tn(x,11) for x in n2],label='N=11')
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
